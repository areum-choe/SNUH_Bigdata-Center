### 1.SBP,Respiration score계산
- SBP / espiration

```python
import pandas as pd
import numpy as np
file='/save/qsofa_file/mimic_SBP,Respiratory.csv'
sbp_respiratory=pd.read_csv(file, sep=',', encoding='CP949')
sbp_respiratory
```

```python
sbp_respiratory= sbp_respiratory.replace({np.nan:None})
```

```python
def qsofa_sbp_respiratory_score(table):

    for i, row in table.iterrows(): 
        #SBP
        if table.at[i,'sbp'] ==  None:
            table.at[i,'SBP_score'] = None
        elif table.at[i,'sbp'] <  90:
            table.at[i,'SBP_score'] =1
        else:
            table.at[i,'SBP_score'] =0

        #Respiratory rate
        if table.at[i,'respiratory_rate'] == None:
            table.at[i,'Respiratory_score'] =None
        elif table.at[i,'respiratory_rate'] >= 22:
            table.at[i,'Respiratory_score'] =1
        else:
            table.at[i,'Respiratory_score'] =0
```

```python
qsofa_sbp_respiratory_score(sbp_respiratory)
```

```python
sbp_respiratory.to_csv(('QSOFA_SCORE_sbp_respiratory.csv'), index=False)
```

### 2.GCS계산
- gcs

```python
file='/save/qsofa_file/mimic_Central nervous system.csv'
gcs=pd.read_csv(file, sep=',', encoding='CP949')
gcs
```

```python
gcs= gcs.replace({np.nan:0})
```

```python
gcs.loc[:,'total_1']=gcs.loc[:,'eye':'motor'].sum(axis=1)
gcs
```

```python
def qsofa_gcs_score(table):

    for i, row in table.iterrows(): 
        if table.at[i,'total_1'] ==  None:
            table.at[i,'gcs_score'] = None
        elif table.at[i,'total_1'] <=  13:
            table.at[i,'gcs_score'] =1
        else:
            table.at[i,'gcs_score'] =0
```

```python
qsofa_gcs_score(gcs)
gcs
```

```python
gcs.to_csv(('QSOFA_SCORE_gcs.csv'), index=False)
```

### QSOFA SCORE 점수 합산

```python
merge_1 = pd.merge(sbp_respiratory[['subject_id','hadm_id','group_hypo','SBP_score','Respiratory_score']]
                   ,gcs[['subject_id','gcs_score']],how='outer',on='subject_id')
merge_1= merge_1.replace({np.nan:0})

merge_1.loc[:,'total'] = merge_1.loc[:,'SBP_score':'gcs_score'].sum(axis=1)
merge_1 = pd.merge(merge_1
                   ,antibio[['subject_id','antibiotic_y']],how='outer',on='subject_id')
merge_1
```

```python
def qsofa_score(table):

    for i, row in table.iterrows(): 
        if table.at[i,'total'] >= 2:
            table.at[i,'qsofa'] =1
        else:
            table.at[i,'qsofa'] =0
```

```python
qsofa_score(merge_1)
merge_1
```

```python
merge_1['qsofa'].value_counts()
```

```python
merge_1.to_csv(('QSOFA_SCORE_TOTAL.csv'), index=False)
```
