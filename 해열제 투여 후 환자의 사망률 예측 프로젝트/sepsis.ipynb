```python
import pandas as pd
# sofa
file='/save/SOFA_SCORE_TOTAL.csv'
SOFA_SCORE_TOTAL=pd.read_csv(file, sep=',', encoding='CP949')

#qsofa
file='/save/QSOFA_SCORE_TOTAL.csv'
QSOFA_SCORE_TOTAL=pd.read_csv(file, sep=',', encoding='CP949')


#icd code
file='/save/qsofa_file/mimic_icdcode_r652제외.csv'
icdcode=pd.read_csv(file, sep=',', encoding='CP949')
```

```python
merge_1 = pd.merge(SOFA_SCORE_TOTAL[['subject_id','hadm_id','sofa']]
                   ,QSOFA_SCORE_TOTAL[['subject_id','qsofa']],how='outer',on='subject_id')

merge_1 = pd.merge(merge_1,icdcode[['subject_id','sepsis']],how='outer',on='subject_id')
```

```python
def sepsis(table):

    for i, row in table.iterrows(): 
        if table.at[i,'sofa'] >= 2 or table.at[i,'sepsis']=='Yes' or table.at[i,'qsofa']>=1:
            table.at[i,'sepsis_final'] =1
        else:
            table.at[i,'sepsis_final'] =0
```

```python
sepsis(merge_1)
merge_1['sepsis_final'].value_counts()
```

```python
merge_1.to_csv(('SEPSIS_FINAL.csv'), index=False)
```
